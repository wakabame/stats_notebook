# 9.1 ベイズの定理
2つの壺 $H_1$, $H_2$ がある. 

$H_1$ には赤玉が $30$ 個, 青玉が $70$ 個ある. 

$H_2$ には赤玉が $70$ 個, 青玉が $30$ 個ある. 

2つの壺のから<u>ランダム</u>に壺を選んで, 12回玉を復元抽出した結果, 赤玉8個, 青玉4個という結果だった. 

## **問題：選ばれた壺は$H_1$, $H_2$ のどちら？**

### 回答

求めるもの：壺 $H_1$ が選ばれたと言える比率。
$$
\dfrac{\textrm{壺 $H_1$ を選び, 抽出結果が赤玉8個, 青玉4個となる確率}}{\textrm{抽出結果が赤玉8個, 青玉4個となる確率}}
$$
ところで, 
\begin{align*}
&\textrm{抽出結果が赤玉8個, 青玉4個となる確率}
\\
 &= \textrm{壺 $H_1$ を選び, 抽出結果が赤玉8個, 青玉4個となる確率} 
\\ 
 &\quad + \textrm{壺 $H_2$ を選び, 抽出結果が赤玉8個, 青玉4個となる確率}   
\end{align*}
である. 


$P(H_1)$：$H_1$の壺を選ぶ確率. 

$A$ : 抽出結果が赤玉8個, 青玉4個となる事象

とすると, 求めるものは
\begin{equation}
\dfrac{P(H_1\cap A)}{P(A)} 
= \dfrac{P(H_1\cap A)}{P(H_1\cap A)+P(H_2\cap A)}
\end{equation}
である. 
さらに条件付き確率
$$
P(H_1 | A) = \dfrac{P(H_1\cap A)}{P(A)} 
$$
という記法を採用すると, 求めるものは

\begin{align}
P(A|H_1)
&= \dfrac{P(H_1\cap A)}{P(A)} \\
&= \dfrac{P(H_1\cap A)}{P(H_1\cap A)+P(H_2\cap A)}\\
&= \dfrac{P(H_1)P(H_1| A)}{P(H_1)P(H_1| A)+P(H_2)P(H_2| A)}.
\end{align}
ここで, ランダムに壺を選んだという仮定から, 
$$\begin{split}
P(H_1\cap A)=\dfrac12 {}_{12} C_8 (0.7)^8(0.3)^4,\\
P(H_2\cap A)=\dfrac12 {}_{12} C_4 (0.7)^4(0.3)^8
\end{split}$$
なので, 

\begin{align}
&\dfrac{P(H_1)P(H_1| A)}{P(H_1)P(H_1| A)+P(H_2)P(H_2| A)}\\
&= \dfrac{\dfrac12 {}_{12} C_8 (0.7)^8(0.3)^4}
{\dfrac12 {}_{12} C_8 (0.7)^8(0.3)^4 +\dfrac12 {}_{12} C_4 (0.7)^4(0.3)^8}\\
&= \dfrac{7^4}{7^4+3^4}\\
&= \dfrac{2401}{2401+81}
\end{align}
思ったより1 に近い.


一方で, $H_2$ が選ばれたと言える確率は, 
$$
\dfrac{81}{2401+81}
$$
である. 

## この計算から感じること
1. 結果から原因の確率分布がわかる
1. ランダムに壺を選んだという事前の設定がある

$P(H_i)$ を事前確率, $P(H_i|A)$ を事後確率という.

# 9.2 事前確率分布と事後確率分布
新しく開発した薬がヒトに効く確率 $\theta_0 \in[0,1]$ を知りたい ⇦ 母数が連続になったよ！

実験データZ:$n$人のうち$z$人には効いた. 

このようなことが起こる確率（**尤度**）は, 
$$
f(z|\theta) = {}_nC_z \theta^z(1-\theta)^{n-z}.
$$
である. 
$$\begin{split}
&\dfrac{\textrm{母数が$\theta$であり, 実験データZが得られる確率}}{\textrm{実験データZが得られる確率}}\\
&=\dfrac{\textrm{母数が$\theta$になる確率}\times \textrm{母数が$\theta$のときに, 実験データZが得られる確率}}
{\textrm{実験データZが得られる確率}}
\end{split}$$
という形で事後分布（$\theta$の確率分布）を求めたい. これを$w'(\theta|z)$と書く. 

とはいえ母数が$\theta$になる確率は不明なので, 事前分布 $w(\theta)$ は自分で決めてしまおう. 

実験データを得るまではまだ何もわからないので, 
$$
w(\theta) = 1 ,\quad \theta \in[0,1]
$$
とする. このとき
$$
w'(\theta|z) = \dfrac{w(\theta)f(z|\theta)}{\int_0^1 w(\theta)f(z|\theta) \,d\theta}
$$
となる. 分母は規格化定数. 
#### 問9.1(ii)のように逐次的に学習をしたい ⇨ 事前分布と事後分布は同じようになってほしい  ⇨ 事前分布はなんでもいいと言ったが, こうした方が便利


では, 実際に実験データ「$5$人のうち$4$人には効いた」が得られた場合を考えよう. 

$\theta \in [0,1],$ に対して, 
$$
w(\theta)f(4|\theta) =5\theta^4(1-\theta)
$$
により, 
$$
\int_0^1 w(\theta)f(4|\theta) \,d\theta = 5B(4,2).
$$
よって事後分布は
$$
w'(\theta|4)
= \dfrac{1}{B(4,2)} \theta^4(1-\theta)
$$
なるベータ分布となっている. 

また, 事後分布に対する $\theta$ の期待値は
$$
\int_0^1 \theta w'(\theta|4) 
=\dfrac{1}{B(4,2)}\int_0^1 \theta\theta^4(1-\theta) \,d\theta = \dfrac{B(5,2)}{B(4,2)}=\dfrac54
$$
となり, 標本平均に一致する.

### [指数型分布族]
実数値確率変数 $X$ が未知パラメータ $\theta$ をもつ(条件付き)確率分布 $p(x|\theta)$に従うとする
(つまり, 観測モデルを $p(x|\theta)$ と定める). 
このとき, $X$ が指数型分布族に属するとは, ある値域の次元が等しい関数 $\phi$, $g$ により, 
\begin{align*}
p(x| \theta) 
&= f(x)\exp\langle \phi(\theta) , g(x)\rangle ,\\
f(x) 
&= \left(\int \exp\langle \phi(\theta) , g(x)\rangle\, d\theta \right)^{-1}
\end{align*}
が成立することをいう. 

### [指数型分布族に対する事前分布]
ハイパーパラメータ $v \in \mathbb{R}^k$ に対して
\begin{align*}
p(\theta| v) 
&= z(v)\exp\langle \phi(\theta) , v\rangle,\\
z(v)
&= \left(\int \exp\langle \phi(\theta) ,v\rangle\, d\theta \ \right)^{-1}
\end{align*}
を事前分布として定める. 
このとき, 
$\hat{v}= \sum_{n=1}^N g(X_n)$ により, 事後分布 $p(\theta|\bar{X})$ は
$$
p(\theta|\bar{X}) = p(\theta|v+\hat{v})
$$
と表せる. すなわち, 事前分布と事後分布における $\phi$ は等しい. 
このように, 分布のうち母数に依存する部分である $\phi$ がベイズ推論によって不変な事前分布を
共役事前分布という. 

(証明)
モデルの設計から, $p(\bar{X},\theta|v)=p(\bar{X}|\theta)p(\theta|v)$ である.
条件付き確率の定義より, 
\begin{align*}
p(\theta|\bar{X},v)
&= \dfrac{p(\bar{X}|\theta,v)}{p(X_\ast|v)}\\
&= \dfrac{p(\bar{X}|\theta)p(\theta|v)}{p(\bar{X}|v)}\\
&= \prod_{n=1}^N p(X_n|\theta)p(\theta|v)/p(\bar{X}|v)\\\
&= \prod_{n=1}^N f(X_n) z(v)p(\bar{X}|v)^{-1} \prod_{n=1}^N\exp\langle \phi(\theta) , g(X_n)\rangle
\exp\langle \phi(\theta) , v\rangle\\
&=z(\hat{v})\exp\langle \phi(\theta) , v+\hat{v}\rangle.
\end{align*}
（終）
## 共役事前分布を用いた予測
事前分布または事後分布が与えられたとき, 次に観測される $X_\ast$ の従う確率分布を予測分布と呼ぶ. 

指数型分布族に対する共役分布については, 次のような公式が与えられる; 

確率変数 $X$ が指数型分布族に属するとし, 共役事前分布 $p(\theta|v)$ をと定められたものとする. 
このとき, 予測分布に対して次が成立する. 
\begin{align*}
p(X_\ast|v)
 &= \dfrac{f(X_\ast)z(v)}{z(v+g(X_\ast))},\\
p(X_\ast|\bar{X},v)
&= \dfrac{f(X_\ast) z(v+\hat{v})}
{z(v+\hat{v}+g(X_\ast))}.
\end{align*}
ここで, 1つ目の式を事前分布に対する予測分布, 2つ目の式を事後分布に対する予測分布と呼ぼう. 

(証明)

事前分布 $p(\theta| v)$ を用いて予測分布 $p(X_\ast|v)$ を求めよう. モデルの設計により, 
$$
p(\theta|X_\ast,v) =\dfrac{p(X_\ast|\theta)p(\theta|v)}{p(X_\ast|v)}
$$
なる関係が成り立つ. これを変形すると, 
\begin{align*}
p(X_\ast|v)
 &=\dfrac{p(X_\ast|\theta)p(\theta|v)}{p(\theta|X_\ast)}\\
 & = \dfrac{f(X_\ast)\exp\langle \phi(\theta),g(X_\ast)\rangle z(v)\exp\langle \phi(\theta) , v\rangle
  }{z(v+g(X_\ast))\exp\langle \phi(\theta),g(X_\ast)+v\rangle}\\
 &= \dfrac{f(X_\ast)z(v)}{z(v+g(X_\ast))}.
\end{align*}
（注：ここでも指数法則が本質的！！）事後分布 $p(\theta|\bar{X})$ を用いて予測分布 $p(X_\ast|\bar{X})$ を求めよう. 
再びモデルの設計により, 
$$
p(\theta|\bar{X},X_\ast,v)=\dfrac{p(X_\ast|\theta)p(\theta|\bar{X},v)}{p(X_\ast|\bar{X},v)}
$$
が成り立つ. 同様の計算により, 
\begin{align*}
p(X_\ast|\bar{X},v)
&= \dfrac{p(X_\ast|\theta)p(\theta|\bar{X},v)}{p(\theta|\bar{X},X_\ast,v)}\\
&= \dfrac{f(X_\ast)\exp\langle\phi(\theta),g(X_\ast)\rangle p(\theta|v+\hat{v})}
{z(v+\hat{v}+g(X_\ast))\exp\{\langle\phi(\theta),g(X_\ast)+v+\hat{v}\rangle\}}\\
&= \dfrac{f(X_\ast)\exp\langle\phi(\theta),g(X_\ast)\rangle z(v+\hat{v})\exp\langle\phi(\theta),v+\hat{v}\rangle}
{z(v+\hat{v}+g(X_\ast))\exp\langle\phi(\theta),g(X_\ast)+v+\hat{v}\rangle}\\
&= \dfrac{f(X_\ast) z(v+\hat{v})}
{z(v+\hat{v}+g(X_\ast))}.
\end{align*}
(終わり)

# 問9.4 ある量 $X$ の母集団分布が $N(\theta,1/4)$ のとき, $\theta$ の事前確率分布として平均$5$, 分散 $1/2$の正規分布 $N(5,1/2)$ を仮定した. $X$ の観測値が $X=6.5$ のとき, $\theta$ の事後確率分布を求めよ. 
### まずは一般に「平均が未知のときのガウス分布」を先ほどの指数型分布族についての議論に持ち込む. 
分散の逆数である精度 $\lambda>0$ は既知として, 平均 $\mu$ が未知の一次元ガウス分布を考える. 
$$
N(x|\mu,\lambda^{-1})
= \sqrt{\dfrac{\lambda}{2\pi}}e^{-\frac{\lambda}{2}x^2}\exp\{-\dfrac{\lambda}{2}\mu^2 +\lambda x\mu\}
$$
により, 
\begin{align*}
f(x)&=\sqrt{\dfrac{\lambda}{2\pi}}e^{-\frac{\lambda}{2}x^2}, \\
\phi_1(\mu)&= -\dfrac{1}{2}\mu^2,\quad \phi_2(\mu) =\mu,\\
g_1(x) &= \lambda, \quad g_2(x)=\lambda x
\end{align*}
なる指数型分布族である. したがって, 事前分布はパラメータ $v=(v_1,v_2)$ より, 
$$
p(\mu|v ) 
=\sqrt{\dfrac{v_1}{2\pi}}e^{-\frac{v_2^2}{2v_1}}\exp\{-\dfrac{v_1}{2}\mu^2 + v_2\mu\}
$$
と置けば事後分布は 
$$
p(\mu|\bar{X})=p(\mu|v+\hat{v})
$$
である. 事前分布は平均が$\frac{v_2}{v_1}$, 精度が$v_1$である正規分布となっているので, 
$a=\frac{v_2}{v_1}$, $b = v_1$ と表すと, 
事前分布は平均$a$, 分散$b$ の正規分布であり, 
\begin{align*}
\hat{a} &= \dfrac{(v+\hat{v})_2}{(v+\hat{v})_1}\\
&= \dfrac{ab+\lambda x}{\hat{b}},\\
\hat{b} &= (v+\hat{v})_1\\
&= b+\lambda
\end{align*}
と置くと, 事後分布は平均$\hat{a}$, 分散$\hat{b}$ の正規分布である. 

$\lambda = 4$, $a=5$, $b=2$を代入すれば,
$$
\hat{b} = 2+4=6,
$$
$$
\hat{a}= \dfrac{5\times 2 + 4\times 6.5}{8}=6
$$
となり, 事後分布は $N(6,1/6)$ である. 

# 9.3 ベイズ推定 & 9.4 統計的決定理論
さっきまでの議論を一旦忘れる。

データから真の母数 $\theta_0$ の推定を行う. 推定とは実現値から母数のよい近似値をみつけること. 
点推定とは違うアプローチを考えよう. 

尤度関数 $f(z|\theta)$ は既知であるものとする. 

すなわち, 母数が $\theta$ であったならば実現値 $z$ が得られる確率が $f(z|\theta)$ という形で得られることを仮定する. 

ここでの目標は, 実現値 $z$ による推定値 $a(z)$ を損失関数の最小化によって構成することだ. 

損失関数の具体的な形は後で与えるとして, $L(\theta-a(z))$ の期待値を下げることを考えていこう. 

まず, $\theta$ を固定する. このとき$z$ の実現確率は $f(z|\theta)$ なので, その期待値は, 
$$
\int f(z|\theta)L(\theta-a(z))\,dz.
$$
母数 $\theta_0$ は未知であるから, 何らかの意味での $\theta$ についての加重平均を取る必要がある. 
これを事前確率$w(\theta)$と呼ぶことにすると, この密度関数による期待値
$$
\int w(\theta)\int f(z|\theta)L(\theta-a(z))\,dzd\theta.
$$
を最小化すればよいことが分かる. 
積分の順序を入れかえることで, 各$z$ について

$$
\int w'(\theta|z)L(\theta-a(z))\,d\theta
$$
を最小化するものが求めるものだとわかる.

```python
df['pred_vi'] = 2*(df['sl']*np.dot(V_inv,mu_vi)[0] + df['sw']*np.dot(V_inv,mu_vi)[1] + df['pl']*np.dot(V_inv,mu_vi)[2] + df['pw']*np.dot(V_inv,mu_vi)[3] ) - np.dot(mu_vi,np.dot(V_inv,mu_vi))
df['pred_se'] = 2*(df['sl']*np.dot(V_inv,mu_se)[0] + df['sw']*np.dot(V_inv,mu_se)[1] + df['pl']*np.dot(V_inv,mu_se)[2] + df['pw']*np.dot(V_inv,mu_se)[3] ) - np.dot(mu_se,np.dot(V_inv,mu_se))
df['pred_ve'] = 2*(df['sl']*np.dot(V_inv,mu_ve)[0] + df['sw']*np.dot(V_inv,mu_ve)[1] + df['pl']*np.dot(V_inv,mu_ve)[2] + df['pw']*np.dot(V_inv,mu_ve)[3] ) - np.dot(mu_ve,np.dot(V_inv,mu_ve))
df['pred'] = df.loc[:,['pred_vi','pred_se','pred_ve']].idxmax(axis=1)

df.head()
```

```python
df['is_correct'] = (df['answer'].str[:2]==df['pred'].str[5:])
df.head()
```
